Problem 2

```python
study_hours = [1, 2, 3, 4, 5]
exam_scores = [50, 55, 65, 70, 80]


print("\nStudy Hours and Exam Scores:")
for hours, score in zip(study_hours, exam_scores):
    print(hours, score)


print("\nInput → Output:")
for hours, score in zip(study_hours, exam_scores):
    print(f"Input: {hours} hours → Output: {score} marks")
```

Problem 3

```python
def pass_fail(score):
    if score >= 60:
        return "Pass"
    else:
        return "Fail"

# Test the function
scores = [45, 67, 82, 59]

for s in scores:
    print(f"Score {s}: {pass_fail(s)}")
```

Problem 4

```python
sizes = [500, 800, 1000, 1200]
prices = [50, 80, 100, 130]

for size, price in zip(sizes, prices):
    print(f"House size: {size} sq ft, Price: {price}")

positive_trend = True
for i in range(1, len(prices)):
    if prices[i] <= prices[i - 1]:
        positive_trend = False
        break

if positive_trend:
    print("Positive trend")
```

Problem 5

```python
study_hours = [2, 4, 6, 8]
scores = [55, 65, 75, 85]

increase_per_2h = (scores[-1] - scores[0]) / ((study_hours[-1] - study_hours[0]) / 2)
predicted_score = scores[-1] + increase_per_2h
print(predicted_score)
```

Problem 6

```python
scores = [35, 58, 62, 74, 89, 45]

for score in scores:
    if score < 50:
        category = "Low"
    elif 50 <= score <= 70:
        category = "Medium"
    else:
        category = "High"

    print(f"Score: {score}, Category: {category}")
```

Problem 7

```python
ages = [18, 20, 22, 24, 26]
heights = [160, 165, 170, 175, 180]

for age, height in zip(ages, heights):
    print(f"Age: {age}, Height: {height} cm")

average_age = sum(ages) / len(ages)
print(f"\nAverage Age: {average_age}")

average_height = sum(heights) / len(heights)
print(f"Average Height: {average_height} cm")

print("Comment: Height appears to increase as age increases in this dataset.")
```
